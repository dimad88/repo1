```python
t=np.linspace(0,10,51)
t
```

```python
f=np.cos(t)
f
```

```python
fig, ax = plt.subplots()

ax.plot(t,f,color = 'green')
ax.grid()

ax.set_xlabel('Значения t')
ax.set_ylabel('Значения f')

ax.set_xlim([0.5 , 9.5])
ax.set_ylim([-2.5 , 2.5])

plt.show()
```

```python
x=np.linspace(-3,3,51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0][0].plot(x,y1)
ax[0][0].set_title('График y1')
ax[0][0].set_xlim([-5, 5])


ax[0][1].plot(x,y2)
ax[0][1].set_title('График y2')

ax[1][0].plot(x,y3)
ax[1][0].set_title('График y3')

ax[1][1].plot(x,y4)
ax[1][1].set_title('График y4')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show()
```
